3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def norm(matrix):
    array = np.array(matrix)
    sm = np.sum(array)
    if sm == 0:
        raise ValueError("Сумма элементов матрицы равна нулю.")
    normalized_array = array / sm
    return normalized_array
print(norm([[1, 2, 3], [4, 5, 6]]))
```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def pyramid_volume(x, y, z):
    x = np.array(x)
    y = np.array(y)
    z = np.array(z)
    product = np.dot(x, np.cross(y, z))
    volume = abs(product) / 6
    return volume
print(pyramid_volume(x=[1, 0, 0], y=[0, 2, 0], z=[0, 0, 4]))
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def create_border_matrix(N):
    matrix = np.zeros((N, N), dtype=int)
        
    matrix[0, :] = 1
    matrix[-1, :] = 1
    matrix[:, 0] = 1
    matrix[:, -1] = 1
    
    return matrix

print(create_border_matrix(8))
```

7. Как найти наиболее близкое значение (к заданному скаляру) в векторе?

```python
def find_closest_value(vector, scalar):
    vector = np.array(vector)    
    
    diff = np.abs(vector - scalar)
    
    index_of_min = np.argmin(diff)
    
    return vector[index_of_min]

print(find_closest_value(vector=[1, 1, 2, 3, 5, 8, 13, 21], scalar=9))
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
def koshi_matrix(X, Y):
    X = np.array(X)
    Y = np.array(Y)
    
    C = 1.0 / (X.reshape(-1,1) - Y)
    
    return C

X = [1, 2, 3]
Y = [4, 5, 6]
print(koshi_matrix(X, Y))
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
import matplotlib.pyplot as plt

merged_data = pd.merge(countries, covid_data, on='location', how='inner')

cases_per_continent = merged_data.groupby('continent')['total_cases'].sum()

plt.figure(figsize=(8, 8))
plt.pie(cases_per_continent, labels=cases_per_continent.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Число случаев COVID-19 по континентам')
plt.axis('equal')
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
merged_data = pd.merge(countries, covid_data, on='location', how='inner')

selected_countries = ['Russia', 'China', 'United States', 'India', 'Nigeria']
covid_tests_data = merged_data[merged_data['location'].isin(selected_countries)]

covid_tests_data = covid_tests_data.sort_values(by='total_tests', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(covid_tests_data['location'], covid_tests_data['total_tests'], color='skyblue')
plt.xlabel('Страны')
plt.ylabel('Число тестов')
plt.title('Число тестов на COVID-19 в России, Китае, США, Индии и Нигерии')
plt.show()
```

4. Постройте гистограмму распределения количества смертей на один миллион человек

```python
merged_data = pd.merge(countries, covid_data, on='location', how='inner')

merged_data['deaths_per_million'] = (merged_data['total_deaths'] / merged_data['population']) * 1e6

merged_data['deaths_per_million'] = merged_data['deaths_per_million'].fillna(merged_data['deaths_per_million'].mean())

plt.figure(figsize=(10, 6))
plt.hist(merged_data['deaths_per_million'], bins=20, color='skyblue', edgecolor='black')

plt.xlabel('Смертность на 1 миллион человек')
plt.ylabel('Частота')
plt.title('Распределение смертности на 1 миллион человек')

max_value = merged_data['deaths_per_million'].max()
step_by_x = 100
plt.xticks(np.arange(0, max_value + step_by_x, step_by_x))

plt.grid(True)
plt.show()
```

6. Сравните распределения кол-ва смертей на 1000 человек в Европе и Африке

```python
merged_data = pd.merge(countries, covid_data, on='location', how='inner')

merged_data['deaths_per_thousand'] = (merged_data['total_deaths'] / merged_data['population']) * 1000

merged_data = merged_data.dropna(subset=['deaths_per_thousand'])

europe_data = merged_data[merged_data['continent'] == 'Europe']
africa_data = merged_data[merged_data['continent'] == 'Africa']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

ax1.hist(europe_data['deaths_per_thousand'], bins=20, color='blue', edgecolor='black')
ax1.set_title('Распределение смертности в Европе')
ax1.set_xlabel('Смертность на 1000 человек')
ax1.set_ylabel('Частота')

ax2.hist(africa_data['deaths_per_thousand'], bins=20, color='green', edgecolor='black')
ax2.set_title('Распределение смертности в Африке')
ax2.set_xlabel('Смертность на 1000 человек')

plt.tight_layout()
plt.show()
```
